```python
def mo(t):
    
    p = []
    for i in range(0,y-u):
        result = (t[i+u]-t[i])/t[i]
        p.append(result)
        if i==(y-u-1):
        
         for j in range(y-u,y):
            t[j] = -1
            p.append(t[j])
            
    return(p)   
```
